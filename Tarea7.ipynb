```python
#Dijkstra
from math import inf 
def getNum(letra):
    return ord(letra)-ord('A')

def getLetra(num):
    return chr(num+ord('A'))

def createGraph(Nodes, Edges):
    graph=[[None]*int(Nodes) for x in range(int(Nodes))]
    for e in Edges:
        i,j,w= e.split()
        i=getNum(i)
        j=getNum(j)
        graph[i][j] = int(w)  # [j,w]
        graph[j][i] = int(w)
        # i1, i2 = [i,j] if i < j else [j,i]
                
    return graph

def dijkstra(graph, start):
    size = len(graph)
    stack = [(start, 0)]
    visited=[False]*size
    distances=[inf]*size
    #path=[]
    while stack:
        n,d = stack.pop()
        if visited[n] and d>=distances[n]:
            continue
        visited[n] = True
        distances[n]=d

        for i in range(size):
            if graph[n][i] is not None:
                stack.append((i,d+graph[n][i]))
        
    return distances #agregar camino

  
#Nodes
Nodes= int(input())
#Begin-End
begin1, end1=input().split()
begin, end=getNum(begin1),getNum(end1)
#edges
Edges= int(input())
#Path
path=[]
nodeN1,nodeN2, Weight=input().split()
for i in range(Edges):
    path.append(input())

#node1,node2 = getNum(nodeN1),getNum(nodeN2)
graph=createGraph(Nodes,Edges)

distances = dijkstra(graph, begin)
#retorna inf? Deberia cambiar inf por "end" de input
#result=map(getLetra,distances)
#print(result)  
```
