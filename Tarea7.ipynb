```python
#Dijkstra
from math import inf 
def getNum(letra):
    return ord(letra)-ord('A')

def getLetra(num):
    return chr(num+ord('A'))

def findLeader(end, start,distances):
    path=""
    while end !=start:
        path += getLetra(end)
        end = distances[end][1]
    path +=getLetra(start)
    distances=distances[end][0]
    return distances,path

def dijkstra(graph, start, end):
    size = len(graph)
    stack = [(start, 0, start )]
    visited=[False]*size
    distances=[inf]*size
    
    while stack:
        n,d,path = stack.pop()
        if visited[n] and d>=distances[n][0]:
            continue
        visited[n] = True 
        distances[n]=(d,path)

        for i in range(size):
            if graph[n][i] is not None:
                stack.append((i,d+graph[n][i],n))
       
    return distances




nodes= int(input())
start,end = input().split()
start =getNum(start)
end = getNum(end)
edges= int(input())

graph=[[None]*int(nodes) for x in range(int(nodes))]

for e in range(int(edges)):
    i,j,w = input().split()
    i =getNum(i)
    j= getNum(j)
    graph[i][j] = int(w)
    #graph[j][i] = int(w)
    

distances = dijkstra(graph, start,end)
distancia,path=findLeader(end,start,distances)
#[getLetra(item) for item in distances]
print(distancia[::-1],path)
```

```python
#Cambio monedas
def cambio(Monedas,Cambio):
    numMonedas = len(Monedas)
    mat=[[0]*(Cambio+1) for x in range(numMonedas)]

    for m in range(numMonedas):
        for c in range(Cambio+1):
            if c == 0:
                mat[m][c]=0
            elif m == 0: # moneda de valor 1
                mat[m][c]=c
            elif Monedas[m]>c:
                mat[m][c]=mat[m-1][c]
            else:
                nuevo = 1 + mat[m][ c-Monedas[m]]
                mat[m][c]= nuevo if nuevo<mat[m-1][c] else mat[m-1][c]

    m=numMonedas-1
    c=Cambio

    while m>0:
        cont=0
        while c>0 and mat[m][c]<mat[m-1][c]:
            cont+=1
            c-=Monedas[m]
        print("De: ${}:{}".format(Monedas[m],cont))
        m-=1
    print("De: ${}:{}".format(Monedas[m],c))
    
    return mat[numMonedas-1][Cambio]
```
