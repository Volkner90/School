```python
import heapq as hq
def getNum(letra):
    return ord(letra)-ord('A')

def getLetra(num):
    return chr(num+ord('A'))


def prim(graph,graph_boolean, n0=0): #change input to recieve two graph and return mst and cost
    mst =[]
    size = len(graph)
    visited =[False]*size
    visited[n0]=True
    pq=[]
    cost = 0

    for i in range(size):
        if graph[n0][i] is not None:
            hq.heappush(pq,(graph[n0][i], n0,i))       
    
    while len(mst)< size-1:
        w,n1,n2 = hq.heappop(pq)
        if visited[n1]!=visited[n2]:
           mst.append((n1,n2,w))
           if not graph_boolean[n1][n2]:
             cost += w
           n = n1 if not visited[n1] else n2
           visited[n]=True
           
           for i in range(size):
                if graph[n][i] is not None and not visited[i]:
                    hq.heappush(pq,(graph[n][i], n,i))
        else: #remove excedent
            if graph_boolean[n1][n2]:
                costExcedent= graph[n1][n2]/ 2   
                cost += costExcedent         
    for i in pq:
        n1= i[1]
        n2= i[2]
        if graph_boolean[n1][n2]:
            costExcedent = graph[n1][n2] / 2
            cost += costExcedent
             
    return mst,cost

nodes, edges= input().split()
nodes = int(nodes)
graph=[[None]*int(nodes) for x in range(int(nodes))]
graph_boolean=[[False]*int(nodes) for x in range(int(nodes))]
for e in range(int(edges)):
    i,j,w,b = input().split()
    i =getNum(i)
    j= getNum(j)
    if b == "T":
       graph_boolean[i][j]= True
       graph_boolean[j][i]= True
    graph[i][j] = int(w)
    graph[j][i] = int(w)
    

mst,cost = prim(graph,graph_boolean)
sum(x[2] for x in mst) 
print(mst) #/1
def column_sum(mst):
      
     return [sum(i) for i in zip(*mst)]

sum=(column_sum(mst)) 
print(float(sum[-1])) 
print(float(cost))  
```
