```python
def getNum(letra):
    return ord(letra)-ord('A')

def getLetra(num):
    return chr(num+ord('A'))

def createRandomGraph(size, vertix):
    graph=[[False]*size for x in range(size)]
    for v in vertix:
        i,j = v.split()
        i =getNum(i)
        j= getNum(j)
        graph[i][j]=True
        graph[j][i]=True
            
    return graph

def findLeader(parents, x):
    while parents[x]!=x:
        x=parents[x]
    return x

def join(parents, rank, i, j):
    leaderI= findLeader(parents,i)
    leaderJ= findLeader(parents,j)

    if leaderI == leaderJ:
        return False
    
    rankI= rank[leaderI]
    rankJ= rank[leaderJ]

    if rankI > rankJ or (rankI == rankJ and leaderI>leaderJ):
        parents[leaderJ]=leaderI
        rank[leaderI]+=rank[leaderJ]
    else:
        parents[leaderI]=leaderJ
        rank[leaderJ]+=rank[leaderI]

    return True

def kruskal(graph):
    edges=[]
    mst=[]
    size=len(graph)
    for i in range(size):
        for j in range(i+1,size):
            if graph[i][j] is not None:
                edges.append((i,j,graph[i][j]))
    
    parents=[*range(size)]
    rank=[1]*size

    edges.sort(key=lambda x:x[2])
    while len(mst)<size-1:
        n1,n2,w = edges.pop(0)
        if join(parents,rank,n1,n2):
            mst.append((n1,n2,w))

    return mst

def connectivityGraph(graph, size):
    connected = True
    subList=[]
    subGraph=[]
    stack=[0]
    while stack:
        CurrentNode = stack.pop()
        for end in range(CurrentNode, size):
            if not DFS(graph, CurrentNode, end):
                connected = False
                stack.insert(0, end)
                continue
            else:
                subGraph.append(getLetra(end))
        if not subArray(subGraph, subList):
            subList.append(subGraph)
        subGraph=[]
    return connected, subList    


inversionistas=9
colaboraciones=1
arr1=[]
for i in range(int(colaboraciones)):
    arr1.append(input())
graph = createRandomGraph(inversionistas,arr1)
print(graph)
mst = kruskal(graph)
total =0
sum(x[2] for x in mst)


connected, subList = connectivityGraph(graph,inversionistas)
if connected:
    print("T")
else:
    print("F")

for subGraph in subList:
    print("".join(subGraph))
```
